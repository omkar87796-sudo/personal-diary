```python
import numpy as np
from scipy.stats import chisquare

# Observed frequencies
observed = np.array([50, 40, 45, 25])

# Expected frequencies (equal distribution)
expected = np.array([40, 40, 40, 40])

# Chi-square goodness of fit
chi_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

# Degrees of freedom
df = len(observed) - 1

alpha = 0.05

# Decision
if p_value < alpha:
    print("Reject Null Hypothesis: Observed distribution is NOT equal.")
else:
    print("Accept Null Hypothesis: Observed distribution fits expected equally.")
```
