```python
import random
import pandas as pd
import numpy as np
```

```python
data = [10,20,30,40,50,60]
```

```python
data = pd.DataFrame(data)
```

```python
data
```

```python
data_mean = data[0].mean()
data_std = data[0].std()
```

```python
def standerd(x):
  return(x-data_mean)/data_std
```

```python
data["standerd"] = data[0].apply(standerd)
```

```python
data
```

```python
def var (x):
  return(x-data_mean)**2
```

```python
data["variance"] = data[0].apply(var)
```

```python
data
```

```python
def normlazation (x):
  return (x-data[0].min())/(data[0].max()-data[0].min())
```

```python
data["normlazation"] = data[0].apply(normlazation)
```
