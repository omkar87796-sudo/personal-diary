```python
import pandas as pd
from scipy import stats
```

```python
data = [47, 49, 48, 46, 50, 49, 47, 48, 46, 47]
data = pd.Series(data)
```

```python
population_mean = 50
alpha = 0.05
```

```python
data_mean = data.mean()
data_std = data.std()
n = len(data)
```

```python
t_stat = (data_mean - population_mean) / (data_std / (n**0.5))
```

```python
t_critical = stats.t.ppf(alpha, df=n-1)
```

```python
if t_stat < t_critical:
    print("Reject the null hypothesis (Left-tailed)")
else:
    print("Fail to reject the null hypothesis (Left-tailed)")
```
