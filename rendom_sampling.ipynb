# without replace

```python
import pandas as pd
import numpy as np
```

we have 100 pepole salary data and we need all pepole avg salary

```python
data = np.random.randint(20000, 80000, 100)
data = pd.DataFrame(data)
```

we made this data using numpy random library

```python
data
```

we put the data in pandas library

```python
group1 = data[0:15]
group2 = data[15:30]
group3 = data[30:45]
group4 = data[45:60]
group5 = data[60:75]
group6 = data[75:90]
group7 = data[90:100]
```

divided in 7 part

```python
total_avrage  = (np.average(group1) + np.average(group2) + np.average(group3) + np.average(group4) + np.average(group5) + np.average(group6) + np.average(group7))/7
print(total_avrage)
```

we take a average of all parts and then add them and divided by 7 and aur final ans is " avrage salary = 51524.019047619055 "

# with replace

```python
data = np.random.randint(20000, 80000, 100)
data = pd.DataFrame(data)
```

```python
data.sample
```

```python
sample1 = data.sample(n=15, replace=True)
sample2 = data.sample(n=15, replace=True)
sample3 = data.sample(n=15, replace=True)
sample4 = data.sample(n=15, replace=True)
sample5 = data.sample(n=15, replace=True)
sample6 = data.sample(n=15, replace=True)
sample7 = data.sample(n=10, replace=True)
```

here we use pandas sample fungtion to the random get random numbers

```python
total_avrage = (np.average(sample1) + np.average(sample2) + np.average(sample3) + np.average(sample4) + np.average(sample5) + np.average(sample6) + np.average(sample7))/7
```
