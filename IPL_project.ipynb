Highlighted batsmen who have already played more than 15 overs and remain consistent for the next 14 overs

```python
batters_15_overs = final_data.groupby("batter")["ball"].count()
batters_15_overs = batters_15_overs[batters_15_overs > 90]
batters_15_overs
```

```python
consistent_batters = []

for batter in batters_15_overs.index:
    batter_data = final_data[final_data["batter"] == batter]

    # Balls faced after the first 15 overs (after 90 deliveries)
    next_phase = batter_data.iloc[90:90+84]     # next 84 balls

    runs = next_phase["batsman_runs"].sum()
    balls = len(next_phase)

    if balls > 0:
        sr = (runs / balls) * 100
    else:
        sr = 0

    # Condition for consistency
    if sr >= 100:        # you can adjust the cutoff
        consistent_batters.append((batter, runs, balls, sr))
```

```python
consistent_df = pd.DataFrame(
    consistent_batters,
    columns=["batter", "runs_next_14_overs", "balls_faced", "strike_rate"]
)

consistent_df.sort_values(by="strike_rate", ascending=False)
```

Analyzed and shortlisted star players who make the most impactful contribution to their teams.

```python
batting = final_data.groupby("batter").agg(
    runs=("batsman_runs", "sum"),
    balls=("batsman_runs", "count"),
).reset_index()

batting["strike_rate"] = (batting["runs"] / batting["balls"]) * 100
batting["impact_batter"] = (
    batting["runs"] * 0.6 +
    batting["strike_rate"] * 0.3 +
    batting["balls"] * 0.1
)
```

```python
bowling = final_data.groupby("bowler").agg(
    balls=("ball", "count"),
    runs=("total_runs", "sum"),
    wickets=("is_wicket", "sum")
).reset_index()

bowling["economy"] = bowling["runs"] / (bowling["balls"] / 6)
bowling["strike_rate"] = bowling["balls"] / bowling["wickets"].replace(0, 1)

bowling["impact_bowler"] = (
    bowling["wickets"] * 0.7 +
    (1 / bowling["economy"]) * 30 +
    (1 / bowling["strike_rate"]) * 40
)
```

```python
batting.rename(columns={"batter": "player"}, inplace=True)
bowling.rename(columns={"bowler": "player"}, inplace=True)

impact_players = pd.merge(
    batting[["player", "impact_batter"]],
    bowling[["player", "impact_bowler"]],
    on="player",
    how="outer"
)

# Fill missing values (some are only batters or only bowlers)
impact_players.fillna(0, inplace=True)

impact_players["total_impact"] = impact_players["impact_batter"] + impact_players["impact_bowler"]

# Sort top stars
top_stars = impact_players.sort_values("total_impact", ascending=False).head(10)

top_stars
```

```python
plt.figure(figsize=(10,6))
plt.bar(top_stars["player"], top_stars["total_impact"])
plt.xticks(rotation=45)
plt.title("Top 10 Impact Players")
plt.ylabel("Impact Score")
plt.tight_layout()
plt.show()
```

```python
# Ensure extras_type is string (fill NaN)
deliveries["extras_type"] = deliveries["extras_type"].fillna("").astype(str).str.lower()

# Ensure extra_runs exists and numeric
deliveries["extra_runs"] = pd.to_numeric(deliveries["extra_runs"], errors="coerce").fillna(0).astype(int)

# Create columns for each extra type by checking extras_type per row
deliveries["is_wide"]     = deliveries["extras_type"] == "wides"
deliveries["is_noball"]   = deliveries["extras_type"] == "noballs"   # some datasets use 'noballs' or 'no-ball' - we handle variants below
deliveries["is_bye"]      = deliveries["extras_type"] == "byes"
deliveries["is_legbye"]   = deliveries["extras_type"] == "legbyes"

# Also handle common alternative spellings/variants just in case
deliveries.loc[deliveries["extras_type"].str.contains("wide", na=False), "is_wide"] = True
deliveries.loc[deliveries["extras_type"].str.contains("noball|no-ball|no ball", na=False), "is_noball"] = True
deliveries.loc[deliveries["extras_type"].str.contains("bye", na=False) & ~deliveries["extras_type"].str.contains("leg", na=False), "is_bye"] = True
deliveries.loc[deliveries["extras_type"].str.contains("legbye", na=False) | deliveries["extras_type"].str.contains("leg-bye", na=False), "is_legbye"] = True

# Sum extras per bowler by type
extras_by_bowler = deliveries.groupby("bowler").agg(
    wides = ("extra_runs", lambda x: x[ deliveries.loc[x.index, "is_wide"] ].sum()),
    noballs = ("extra_runs", lambda x: x[ deliveries.loc[x.index, "is_noball"] ].sum()),
    byes = ("extra_runs", lambda x: x[ deliveries.loc[x.index, "is_bye"] ].sum()),
    legbyes = ("extra_runs", lambda x: x[ deliveries.loc[x.index, "is_legbye"] ].sum()),
    total_extra_runs = ("extra_runs", "sum"),
    total_deliveries = ("ball", "count")   # total rows (deliveries) attributed to that bowler
).reset_index()

# Compute legal deliveries per bowler:
# wides and no-balls are NOT legal deliveries (they do not consume a legal ball),
# but dataset rows for wides/noballs may still be present. We'll treat a delivery as legal
# if it's NOT classified as wide or no-ball.
deliveries["is_legal"] = ~(deliveries["is_wide"] | deliveries["is_noball"])

legal_balls = deliveries[deliveries["is_legal"]].groupby("bowler")["ball"].count().rename("legal_deliveries").reset_index()

# Merge legal balls into extras_by_bowler
extras_by_bowler = extras_by_bowler.merge(legal_balls, on="bowler", how="left")
extras_by_bowler["legal_deliveries"] = extras_by_bowler["legal_deliveries"].fillna(0).astype(int)

# Compute overs bowled (using legal deliveries)
extras_by_bowler["overs_bowled"] = extras_by_bowler["legal_deliveries"] / 6.0

# Avoid divide-by-zero when computing extras per over
extras_by_bowler["extras_per_over"] = np.where(
    extras_by_bowler["overs_bowled"] > 0,
    extras_by_bowler["total_extra_runs"] / extras_by_bowler["overs_bowled"],
    np.nan
)

# Sort by total extras (highest offenders)
top_extra_bowlers = extras_by_bowler.sort_values(by="total_extra_runs", ascending=False).head(5)

# Nicely format and show results
cols = ["bowler", "wides", "noballs", "byes", "legbyes", "total_extra_runs", "legal_deliveries", "overs_bowled", "extras_per_over"]
print(top_extra_bowlers[cols].to_string(index=False))
```

# Fielder Selection Analysis

```python
# Clean column names
deliveries.columns = deliveries.columns.str.strip()
matches.columns = matches.columns.str.strip()

# ---------- Merge deliveries + matches ----------
merged_df = deliveries.merge(
    matches[["id", "season"]],
    left_on="match_id",
    right_on="id",
    how="left"
)

# ---------- Extract only catches ----------
# dismissal_kind = 'caught' â†’ indicates catch taken
catches = merged_df[merged_df["dismissal_kind"] == "caught"]

# fielder column â†’ contains the player who took the catch
catches = catches[catches["fielder"].notna()]

# ---------- Total catches by each fielder ----------
fielder_catch_count = catches.groupby("fielder").size().reset_index(name="total_catches")

# ---------- Matches played as fielder ----------
matches_fielded = catches.groupby("fielder")["match_id"].nunique().reset_index(name="matches_fielded")

# ---------- Merge to compute consistency ----------
fielder_stats = fielder_catch_count.merge(matches_fielded, on="fielder")

# Catches per match = consistency indicator
fielder_stats["catches_per_match"] = fielder_stats["total_catches"] / fielder_stats["matches_fielded"]

# ---------- Top 10 best fielders ----------
top_fielders = fielder_stats.sort_values(
    by=["total_catches", "catches_per_match"],
    ascending=[False, False]
).head(10)

print("ðŸŽ¯ Top 10 Fielders With Best Catching Performance:")
print(top_fielders)

# ---------- Visualization ----------
plt.figure(figsize=(10, 5))
plt.bar(top_fielders["fielder"], top_fielders["total_catches"])
plt.xlabel("Fielder")
plt.ylabel("Total Catches")
plt.title("Top 10 Fielders With Highest Number of Catches")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Wicket Keeper Analysis

```python
# Clean column names
deliveries.columns = deliveries.columns.str.strip()
matches.columns = matches.columns.str.strip()

# ---------- Merge Deliveries + Matches for Match Information ----------
merged_df = deliveries.merge(
    matches[["id", "season"]],
    left_on="match_id",
    right_on="id",
    how="left"
)

# ---------- Filter dismissals where keeper is involved ----------
# Wicket keeper actions = caught behind + stumping
wk_actions = merged_df[
    merged_df["dismissal_kind"].isin(["caught", "stumped"])
]

# Must have player involved in fielding â†’ 'fielder'
wk_actions = wk_actions[wk_actions["fielder"].notna()]

# ---------- Count dismissals for each wicket keeper ----------
wk_dismissals = wk_actions.groupby("fielder").size().reset_index(name="total_dismissals")

# ---------- Count matches where each keeper participated ----------
wk_matches = wk_actions.groupby("fielder")["match_id"].nunique().reset_index(name="matches_as_keeper")

# ---------- Merge final stats ----------
wk_stats = wk_dismissals.merge(wk_matches, on="fielder")

# Add consistency score
wk_stats["dismissals_per_match"] = wk_stats["total_dismissals"] / wk_stats["matches_as_keeper"]

# ---------- Sort top 10 wicket keepers ----------
top_wk = wk_stats.sort_values(
    by=["total_dismissals", "dismissals_per_match"],
    ascending=[False, False]
).head(10)

print("ðŸŽ¯ Top Wicket Keepers Based on Overall Performance")
print(top_wk)

# ---------- Visualization ----------
plt.figure(figsize=(10, 5))
plt.bar(top_wk["fielder"], top_wk["total_dismissals"])
plt.title("Top Wicket Keepers by Total Dismissals")
plt.xlabel("Wicket Keeper")
plt.ylabel("Total Dismissals")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
