# T-Test in Hypothesis Testing

## Left test

```python
import pandas as pd
from scipy import stats
```

```python
data = [47, 49, 48, 46, 50, 49, 47, 48, 46, 47]
data = pd.Series(data)
```

```python
population_mean = 50
alpha = 0.05
```

```python
data_mean = data.mean()
data_std = data.std()
n = len(data)
```

```python
t_stat = (data_mean - population_mean) / (data_std / (n**0.5))
```

```python
t_critical = stats.t.ppf(alpha, df=n-1)
```

```python
if t_stat < t_critical:
    print("Reject the null hypothesis (Left-tailed)")
else:
    print("Fail to reject the null hypothesis (Left-tailed)")
```

## Right test

```python
data = [47, 49, 48, 46, 50, 49, 47, 48, 46, 47]
data = pd.Series(data)
```

```python
population_mean = 50
alpha = 0.05
```

```python
data_mean = data.mean()
data_std = data.std()
n = len(data)
```

```python
t_stat = (data_mean - population_mean) / (data_std / (n**0.5))
```

```python
t_critical = stats.t.ppf(1-alpha, df=n-1)
```

```python
if t_stat > t_critical:
    print("Reject the null hypothesis (Left-tailed)")
else:
    print("Fail to reject the null hypothesis (Left-tailed)")
```

## Two tailed independed

```python
import numpy as np
```

```python
group1 = np.array([18, 14, 24, 26, 22, 34, 25, 24, 27, 19])
group2 = np.array([15, 12, 21, 24, 20, 30, 23, 22, 25, 18])
```

```python
n1 = len(group1)
n2 = len(group2)

mean1 = group1.mean()
mean2 = group2.mean()

std1 = group1.std(ddof=1)
std2 = group2.std(ddof=1)
```

```python
sp = np.sqrt(((n1-1)*std1**2 + (n2-1)*std2**2) / (n1 + n2 - 2))
```

```python
t_stat = (mean1 - mean2) / (sp * np.sqrt(1/n1 + 1/n2))
```

```python
alpha = 0.05
df = n1 + n2 - 2

t_critical_right = stats.t.ppf(1 - alpha/2, df)
t_critical_left = stats.t.ppf(alpha/2, df)
```

```python
if t_stat > t_critical_right or t_stat < t_critical_left:
    print("Reject Null Hypothesis (Two-tailed Independent t-test)")
else:
    print("Accept Null Hypothesis (Two-tailed Independent t-test)")
```
