```python
import pdfplumber
from gtts import gTTS

pdf_path = "/content/The_Silver_Key_Story.pdf" #add her yoy pdf path

text = ""
with pdfplumber.open(pdf_path) as pdf:
    for page in pdf.pages:
        page_text = page.extract_text()
        if page_text:
            text += page_text

if text.strip() == "":
    print("pdf text is not readeble format")
else:
    tts = gTTS(text)
    tts.save("your ready audio.mp3") # here you can rename your file
    print("Download it from the file.")
```
