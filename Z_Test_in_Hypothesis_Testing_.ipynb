## Two tailed test



```python
import pandas as pd
from scipy import stats
```

```python
data = [52, 48, 47, 51, 49, 50, 53, 46, 54, 49]
data = pd.DataFrame(data)
```

```python
len_of_data = len(data)
population_mean = 50
std_population = 3
alpha = 0.05
```

```python
sample_mean = data[0].mean()
```

```python
z_sample = (sample_mean - population_mean) / (std_population / (len_of_data ** 0.5))
```

```python
z_critical_right = stats.norm.ppf(1 - alpha/2)
z_critical_left = -z_critical_right
```

```python
if z_sample < z_critical_left or z_sample > z_critical_right:
    print(" Reject Null Hypothesis ")
else:
    print(" Fail to Reject Null Hypothesis ")
```

## Left tailed test

```python
import math
```

```python
data = [47, 49, 48, 46, 50, 49, 47, 48, 46, 47]
data = pd.Series(data)

# Known values
population_mean = 50
std_population = 3
alpha = 0.05
```

```python
sample_mean = data.mean()
n = len(data)
standard_error = std_population / math.sqrt(n)
```

```python
z_value = (sample_mean - population_mean) / standard_error
```

```python
z_critical = stats.norm.ppf(alpha)
```

```python
if z_value < z_critical:
    print(" Reject Null Hypothesis ")
else:
    print(" Fail to Reject Null Hypothesis ")
```

## Right tailed test

```python
data = [47, 49, 48, 46, 50, 49, 47, 48, 46, 47]
data = pd.Series(data)

# Known values
population_mean = 50
std_population = 3
alpha = 0.05
```

```python
len_data = len(data)
population_mean = 50
std_population = 3
alpha = 0.05
```

```python
sample_mean = data[0].mean()
```

```python
z_sample = (sample_mean - population_mean) / (std_population / (len_data ** 0.5))
```

```python
z_critical = stats.norm.ppf(1 - alpha)
```

```python
if z_sample > z_critical:
    print(" Reject Null Hypothesis ")
else:
    print(" Fail to Reject Null Hypothesis ")
```
