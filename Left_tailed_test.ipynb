```python
import pandas as pd
from scipy import stats
import math
```

```python

data = [47, 49, 48, 46, 50, 49, 47, 48, 46, 47]
data = pd.Series(data)

# Known values
population_mean = 50
std_population = 3
alpha = 0.05
```

```python
sample_mean = data.mean()
n = len(data)
standard_error = std_population / math.sqrt(n)
```

```python
z_value = (sample_mean - population_mean) / standard_error
```

```python
z_critical = stats.norm.ppf(alpha)
```

```python
print("Sample Mean:", sample_mean)
print("Z-Statistic:", z_value)
print("Left Critical Z:", z_critical)

```

```python

if z_value < z_critical:
    print(" Reject Null Hypothesis ")
else:
    print(" Fail to Reject Null Hypothesis ")
```
