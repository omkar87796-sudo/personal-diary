```python
import numpy as np
from scipy import stats
```

```python
before = np.array([18, 14, 24, 26, 22, 34, 25, 24, 27, 19])
after  = np.array([15, 12, 21, 24, 20, 30, 23, 22, 25, 18])
```

```python
d = before - after
n = len(d)
```

```python
d_mean = d.mean()
d_std  = d.std(ddof=1)
```

```python
t_data = d_mean / (d_std / np.sqrt(n))
```

```python
alpha = 0.05
df = n - 1
```

```python
t_critical_right = stats.t.ppf(1 - alpha/2, df)
t_critical_left  = stats.t.ppf(alpha/2, df)
```

```python
if t_data > t_critical_right or t_data < t_critical_left:
    print("Reject Null Hypothesis")
else:
    print("Accept Null Hypothesis")
```
